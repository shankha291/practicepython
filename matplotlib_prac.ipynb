```python
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
x=[1,2,3,4]
y=[3,6,9,12]
style1=dict(
    marker=".",
    mfc="red", 
    mec="blue", 
    ms="20", 
    #line styles
    ls="dashed", 
    lw= 2, 
    color="green"
)
style2=dict(
    fontsize=15, 
    family="arial", 
    color="red", 
    fontweight="bold"
)
plt.plot(x,y,**style1)
plt.title("Matplotlib",**style2)
plt.xlabel("x-axis",**style2)
gridstyle=dict(
    lw="2", 
    color="gray", 
    ls="solid"
)
plt.grid(axis="x", 
        **gridstyle)
plt.grid(axis="y", 
        **gridstyle)
plt.ylabel("y-axis",**style2)
plt.show()
```

```python
#+++++++Barchart++++++++++
category=["c1","c2","c3","c4","c5"]
values=[4,3,5,4,1]
color1=["red","blue","green","brown","skyblue"]
plt.title("Bar chat",**style2)

plt.bar(category,values,color=color1)
#for horizontal
# plt.barh(category,values)

plt.show()
```

```python
plt.pie(values, 
       labels=category, 
       autopct="%1.1f%%", 
       colors=color1, 
       explode=[0,0,0.1,0,0], 
        shadow=True, 
        
       )
plt.show()
```
